### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
```

```python
url='https://www.naukri.com/'
driver.get(url)
```

```python
search_button=driver.find_element_by_id('qsb-keyword-sugg')
search_button.send_keys('Data Analyst')

search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')


search=driver.find_element_by_xpath("//button[@class='btn']")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
exp_req=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    job_title.append(i.text)
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span"):
    job_location.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='mt-7 companyInfo subheading lh16']//a"):
    company_name.append(i.text)
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span"):
    exp_req.append(i.text)
df=pd.DataFrame({})
df['Job_title']=job_title[:10]
df['job_location']=job_location[:10]
df['company']=company_name[:10]
df['experience']=exp_req[:10]
df
```

### Q4: Write a python program to scrape data for first 10 job results for Data scientist Designation in Noida location. You have to scrape company_name, No. of days ago when job was posted, Rating of the company.

```python
url='https://www.glassdoor.co.in/index.htm'
driver.get(url)
```

```python
search_job1 = driver.find_element_by_id('sc.keyword')

search_job1.send_keys('Data Scientist')

search_job2 = driver.find_element_by_id('sc.location')

driver.find_element_by_id('sc.location').clear()

search_job2.send_keys('Noida')
```

```python
search_btn = driver.find_element_by_xpath("//button[@class='gd-ui-button ml-std col-auto SearchStyles__newSearchButton css-iixdfr']")

search_btn.click()
```

```python
import time
from selenium.common.exceptions import NoSuchElementException
# getting all the URL's

jobs = driver.find_elements_by_xpath("//a[@class='jobLink css-1rd3saf eigr9kq2']")

url=[]

for i in jobs:

    url.append(i.get_attribute('href'))
# gathering the data

comp=[]

rating=[]

days=[]

d = driver.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-mi55ob']")

for i in d[:10]:

    days.append(i.text)
    
for ur in url:

    driver.get(ur)

    time.sleep(2)

    comp.append(driver.find_element_by_xpath("//div[@class='css-16nw49e e11nt52q1']").text[:-8])

    try:

        rating.append(driver.find_element_by_xpath("//span[@class='css-1pmc6te e11nt52q4']").text[:-3])

    except NoSuchElementException:

        rating.append('-')
#slicing to get top 10 rows

comp = comp[:10]

rating = rating[:10]
#creating the dataframe

jobs_d = pd.DataFrame({

             "Company_Name": comp,

             "rating": rating,

             "days ago":days})
jobs_d
```

```python
company = driver.find_elements_by_xpath("//a[@class='css-f3vw95 e1aj7ssy3']")

nsalary = driver.find_elements_by_xpath("//div[@class='col-12 col-lg-auto']")

rating = driver.find_elements_by_xpath("//span[@class='m-0 css-kyx745']")

avgsalary = driver.find_elements_by_xpath("//div[@class='col-12 col-lg-4 px-lg-0 d-flex align-items-baseline']")

minsalary = driver.find_elements_by_xpath("//div[@class='d-flex mt-xxsm css-79elbk epuxyqn0']")




#converting data into suitable forms

comp =[]

nsal=[]

rat=[]

avgsal=[]

minsal=[]

maxsal=[]

for i in company[:10]:

    comp.append(i.text)

for i in nsalary[:10]:

    nsal.append(i.text)

for i in rating[:10]:

    rat.append(i.text)

for i in avgsalary[:10]:

    avgsal.append(i.text)

for i in minsalary[:10]:

    st = i.text.split("\n")

    minsal.append(st[0])

    maxsal.append(st[1])
jobs_data = pd.DataFrame({"Company": comp,

             "number fo salaries": nsal,

             "rating": rat,

             "average salary":avgsal,

             "minimum salary": minsal,

             "maximum salary": maxsal,})
jobs_data

```

```python
apply=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']")
apply.click()
```

```python
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text=='Intel Core i7':
        i.click()
        break
```

```python
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text=='Intel Core i9':
        i.click()
        break
```

```python
Title=[]
Ratings=[]
price=[]
titles=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
for i in titles[:10]:
    Title.append(i.text)
prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in prices[:10]:
    price.append(i.text)
```

```python
from selenium.common.exceptions import NoSuchElementException
urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
UR=[]
for i in urls[:10]:
    UR.append(i.get_attribute('href'))
for url in UR:
    driver.get(url)
    try:                  
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()                                                      
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings.append(rating.text)
        
    except NoSuchElementException   as e:
        Ratings.append("NO rating")
```

```python
df=pd.DataFrame({'Title':Title,
                'Price':price,
                'Ratings':Ratings})

df
```
