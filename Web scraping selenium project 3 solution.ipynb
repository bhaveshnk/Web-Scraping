### Q.1 Write a python program which searches all the product under a particular product from
www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user
input is ‘guitar’. Then search for guitars.

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.amazon.in/'
driver.get(url)
```

```python
product = input('Enter the product name: ')
type_button = driver.find_element_by_id('twotabsearchtextbox')
type_button.clear()
type_button.send_keys(product)

search = driver.find_element_by_id('nav-search-submit-button')
search.click()
```

### Q2-In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“

```python
url = []

for i in range(1, 3):
    url_tags = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in url_tags:
        url.append(i.get_attribute('href'))
    
    next_button = driver.find_element_by_xpath("//li[@class='a-last']/a")
    nxt = next_button.get_attribute('href')
    driver.get(nxt)
```

```python
brand=[]
product=[]
rating=[]
total_rating=[]
availability=[]
price=[]
delivery=[]
replacement=[]
other=[]
URL=[]

for ur in url:
    driver.get(ur)
    try:
        brand_tag=driver.find_element_by_xpath("//a[@id='bylineInfo']")
        brand.append(brand_tag.text)
    except NoSuchElementException:
        brand.append("--")
        
    try:
        product_tag=driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']/span")
        product.append(product_tag.text)
    except NoSuchElementException:
        product.append('--')
        
    try:
        rating_tag=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(rating_tag.text)
    except NoSuchElementException:
        rating.append('--')
        
    try:
        total_tag=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        total_rating.append(total_tag.text)
    except NoSuchElementException:
        total_rating.append('--')
        
    try:    
        availability_tag=driver.find_element_by_xpath("//div[@id='availability']/span")
        availability.append(availability_tag.text)
    except NoSuchElementException:
        availability.append('--')
        
    try:
        price_tag=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        price.append(price_tag.text)
    except NoSuchElementException:
        price.append('--')
        
    try:
        delivery_tag=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b")
        delivery.append(delivery_tag.text)
    except NoSuchElementException:
        delivery.append('--')
        
    try:
        replacement_tag=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        replacement.append(replacement_tag.text)
    except NoSuchElementException:
        replacement.append('--')
        
    try: 
        other_tag=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
        other.append(other_tag.text)
    except NoSuchElementException:
        other.append('--')
        
    URL.append(ur) 
    time.sleep(5)
```

```python
df=pd.DataFrame({'Brand':brand,
                 'Product':product,
                 'Price':price,
                 'Rating':rating,
                 'Total_Rating':total_rating,
                 'Availability':availability,
                 'Delivery':delivery,
                 'Replacement':replacement,
                 'Other_detail':other,
                 'URL':URL})
```

### Q.5 Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver = webdriver.Chrome('chromedriver.exe')

url = 'https://www.google.com/maps'
driver.get(url)
```

```python
def get_long_lat_gmap_for(city):
    driver.find_element_by_id("searchboxinput").send_keys(city)
    driver.find_element_by_id("searchbox-searchbutton").click()
    
    try:
        element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, "//button[@data-value='Directions']"))
        )
    finally: 
        cur_url = driver.current_url

    cur_url = cur_url.split('@')[1].split(',')
    driver.quit()
    return "Longitude: "+cur_url[0],"Latitude: "+cur_url[1]

    
```

```python
get_long_lat_gmap_for('new york')
```

```python
urls = []
name =[]
distance = []
ratings = []
total_reviews = []
price = []
private = []
dorms = []
facilities = []
property_description = []

for i in driver.find_elements_by_xpath("//h2[@class='title title-6']/a"):
    urls.append(i.get_attribute('href'))

for i in driver.find_elements_by_xpath("//span[@class='description']"):
    distance.append(i.text)
try:    
    for i in driver.find_elements_by_xpath("//div[@class='score orange big']"):
        ratings.append(i.text)
except NoSuchElementException:
    ratings.append('--')
    
for i in driver.find_elements_by_xpath("//div[@class='reviews']"):
    total_reviews.append(i.text)

for i in driver.find_elements_by_xpath('//div[@class="price-col"]'):
    price.append(i.text)

for i in price:
    if 'Privates' in i:
        private.append(i)
    else:
        dorms.append(i)
    
for url in urls:
    driver.get(url)
    time.sleep(4)
    for i in driver.find_elements_by_xpath("//div[@class='title-2']"):
        name.append(i.text)
        
    for i in driver.find_elements_by_xpath('//ul[@class="facilities"]'):
        facilities.append(i.text)
    
    for i in driver.find_elements_by_xpath("//div[@class='content collapse-content']"):
        property_description.append(i.text)
        
```
